```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing(as_frame=True)
```

```python
print(california_housing)
```

```python
# loading data as pandas data frame
house_data = pd.DataFrame(california_housing.data, columns = california_housing.feature_names)
```

```python
# print first five rows of the datasets
house_data.head()
```

```python
# add target column to the dataFrame
house_data['MedHouseVal'] = california_housing.target
```

```python
# print first five rows of the datasets
house_data.head()
```

```python
house_data.shape
```

```python
house_data.isna().sum()
```

```python
house_data.describe()
```

understanding the correletion between features of dataset

```python
correlation = house_data.corr()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square = True, fmt = '.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
X = house_data.drop(columns = ['MedHouseVal'], axis = 1)
Y = house_data['MedHouseVal']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

Model trainig

```python
# loading the model
model = XGBRegressor()
```

```python
# Training the the model
model.fit(X_train, Y_train)
```

```python
# prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R square error
r2_train = metrics.r2_score(Y_train, training_data_prediction)
# mean absolute error
mae = metrics.mean_absolute_error(Y_train, training_data_prediction)
print("R square error: ", r2_train )
print("Mean: ", mae)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual price vs Predicted")
plt.show()
```
