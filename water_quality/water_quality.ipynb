```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
water_data = pd.read_csv('/content/water_potability.csv')
```

```python
# analize the data
water_data.head()
```

```python
water_data.shape
```

```python
water_data.isna().sum()
```

```python
water_data.dropna()
```

```python
water_data.isna().sum()
```

```python
print(water_data['ph'].mean())
```

```python
water_data['ph'] = water_data['ph'].fillna(water_data['ph'].mean())
```

```python
water_data.isna().sum()
```

```python
water_data['Sulfate'] = water_data['Sulfate'].fillna(water_data['Sulfate'].mean())
```

```python
water_data['Trihalomethanes'] = water_data['Trihalomethanes'].fillna(water_data['Trihalomethanes'].mean())
```

```python
water_data.isna().sum()
```

```python
X = water_data.drop('Potability', axis = 1)
Y = water_data['Potability']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)
```

```python
# accuracy on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```
